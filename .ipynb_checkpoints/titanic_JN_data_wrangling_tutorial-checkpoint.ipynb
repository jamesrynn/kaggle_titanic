# 1. Load Packages and Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
import os
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

%matplotlib inline
rcParams['figure.figsize'] = 10,8
sns.set(style='whitegrid', palette='muted',
        rc={'figure.figsize': (12,8)})
```

```python
# Load data as Pandas dataframe
train = pd.read_csv('./data/train.csv')
test = pd.read_csv('./data/test.csv')
df = pd.concat([train, test], axis=0, sort=True)

# Display combined data values.
df.head()
```

```python
# Display statistics of the data.
def display_all(df):
    with pd.option_context("display.max_rows", 1000, "display.max_columns", 1000): 
        display(df)

        
display_all(df.describe(include='all').T)
```

```python
# Count how many people dies and how many survived.
df['Survived'].value_counts()
```


# 2. IMPUTATION

## 2.1 Impute Missing Age Values

Impute missing ages using median of complete sets with same title.

```python
# Create new title column for all data sets.
df['Title'] = df['Name'].str.extract('([A-Za-z]+)\.', expand=True)
df.head()

train['Title'] = train['Name'].str.extract('([A-Za-z]+)\.', expand=True)

test['Title'] = test['Name'].str.extract('([A-Za-z]+)\.', expand=True)
```

Use only most common title by finding most frequent and grouping.

```python
df['Title'].value_counts()
```

```python
# Replace less frequent titles with equivalent/more common ones.

mapping = {'Col' : 'Mr',
           'Mlle' : 'Miss',
           'Major' : 'Mr',
           'Ms' : 'Miss',
           'Jonkheer' : 'Mr',
           'Capt' : 'Mr',
           'Mme' : 'Mrs',
           'Dona' : 'Mrs',
           'Don' : 'Mr',
           'Countess' : 'Mrs',
           'Sir' : 'Mr',
           'Lady' : 'Mrs'}

df.replace({'Title' : mapping}, inplace=True)
df['Title'].value_counts()

train.replace({'Title' : mapping}, inplace=True)
test.replace({'Title' : mapping}, inplace=True)
```

```python
# Use median of group with same title to impute the missing ages (using training data only).

title_ages = dict(train.groupby('Title')['Age'].median())
print(title_ages)
```

```python
# create a column of the median ages
df['age_med'] = df['Title'].apply(lambda x: title_ages[x])

# replace the missing ages (for ALL data) with the value in the median ages column
df['Age'].fillna(df['age_med'], inplace=True)

# delete median age column 
del df['age_med']
```

```python
# Visualise median ages for each title used.
sns.barplot(x='Title', y='Age', data=train, estimator=np.median, ci=None, palette='Blues_d')
plt.xticks(rotation=45)
plt.show()
```

```python
# visualise survival by title
sns.countplot(x='Title', data=train, palette='hls', hue='Survived')
plt.xticks(rotation=45)
plt.show()
```

## 2.2 Impute Missing Fare Values

```python
# Visualise how the fare affects survival

sns.swarmplot(x='Sex', y='Fare', hue='Survived', data=train)
plt.show()
```

There is a single fare value missing, use the median fare value (of training data) for the passengers class to impute value.

```python
# median of fare cost fpr each passenger class group 
class_fares = dict(train.groupby('Pclass')['Fare'].median())

# created column of average fares
df['fare_med'] = df['Pclass'].apply(lambda x: class_fares[x])

# replace all missing fares with the average and delete unneccessary column
df['Fare'].fillna(df['fare_med'], inplace=True)
del df['fare_med']
```

## 2.3 Impute Missing 'Emarked' Value

```python
# visualise survival rates for each point of embarkment
sns.catplot(x='Embarked', y='Survived', data=train, kind='bar', palette='muted', ci=None)
plt.show()
```

```python
# use Pandas inbuilt 'backfill' method to impute the (two) missing values
df['Embarked'].fillna(method='backfill', inplace=True)
```

### Add 'Family Size' column

```python
# create 'Family_Size' column (Parch + SibSp + self)
df['Family_Size'] = df['Parch'] + df['SibSp'] + 1

display_all(df.describe(include='all').T)
```

# Save cleaned version

```python
# redefine training and test data to be imputed versions
train = df[pd.notnull(df['Survived'])]
test = df[pd.isnull(df['Survived'])]

# save clean version to csv files
train.to_csv('./data/train_clean.csv', index=False)
test.to_csv('./data/test_clean.csv', index=False)
```
